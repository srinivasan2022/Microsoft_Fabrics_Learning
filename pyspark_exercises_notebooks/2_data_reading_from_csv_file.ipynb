### Read the data from CSV file
**Let's upload the IPL cricket CSV file**

### 

```python
from pyspark.sql import SparkSession

file_path = "Files/data/ipl_summary_raw.csv"

df = spark.read.csv(file_path , header=True , inferSchema=True)

df.show(5)
```

**Head()**

```python
display(df.head(10))
```

**printSchema()**

```python
df.printSchema()
```

### Basic Operation
**Select the specific columns**

```python
# Select 
df.select("info_city" , "info_teams_1" , "info_teams_2").show()
```

```python
dis = df.select("info_city" , "info_teams_1" , "info_teams_2")
display(dis)
```

**Show distinct values of a column**

```python
# distinct 
df.select("info_city").distinct().show()
```

```python
df.select("info_teams_1").distinct().show()
```

**Filter rows based on a condition**

```python
# Filter
df.filter(df["info_city"] == "Chennai").show()
```

```python
filter_rows = df.filter(df["info_city"] == "Chennai")
display(filter_rows)
```
