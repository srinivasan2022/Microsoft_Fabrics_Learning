### Let's upload the Supermart Grocery Sales - Retail Analytics Dataset

**Read the CSV file**

```python
df = spark.read.format("csv").option("header","true").load("Files/data/Supermart Grocery Sales - Retail Analytics Dataset.csv")
display(df)
```

### Creates the temporary SQL table

```python
df.createOrReplaceTempView("orders")
```

### Select columns

```python
%%sql
SELECT Order_ID, Customer_Name, Sales 
FROM orders;
```

### Filter - where clause

```python
%%sql
SELECT * 
FROM orders 
WHERE City = 'Madurai';
```

### OrderBy function

```python
%%sql
SELECT * 
FROM orders 
ORDER BY Sales DESC;
```

### GroupBy function

```python
%%sql

SELECT Region, COUNT(*) AS Total_Orders 
FROM orders 
GROUP BY Region;
```

### distinct - remove duplicate values

```python
%%sql
SELECT DISTINCT City
FROM orders;
```

### count , where and GroupBy

```python
%%sql
SELECT city, COUNT(*) AS Total_Orders 
FROM orders 
WHERE Category = 'Snacks'
GROUP BY city ;
```

```python
%%sql
SELECT * 
FROM orders 
WHERE Discount > 0.3;
```

```python
%%sql
SELECT * 
FROM orders 
WHERE Sales > 1000 AND Profit > 200;
```

### Renaming the column name

```python
%%sql
SELECT Order_ID AS OrderNumber, Customer_Name AS Customer, Sales AS Revenue 
FROM orders;
```

### Like operator

```python
%%sql
SELECT * 
FROM orders 
WHERE Customer_Name LIKE '%Ram%';
```

### Isin or contains

```python
%%sql
SELECT * 
FROM orders 
WHERE City IN ('Madurai', 'Perambalur');
```

```python
%%sql
SELECT * 
FROM orders 
ORDER BY Profit ASC, Sales DESC;
```

### Average

```python
%%sql
SELECT AVG(Sales) AS Average_Sales 
FROM orders;
```

### Replace values

```python
%%sql
SELECT Order_ID, 
       REPLACE(State, 'Tamil Nadu', 'TN') AS State_Abbreviation 
FROM orders;
```
